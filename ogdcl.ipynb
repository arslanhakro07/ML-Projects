```python
# prompt: Write code to perform time series cross-validation (e.g., with TimeSeriesSplit) to check the generalizability of the RNN model.

from sklearn.model_selection import TimeSeriesSplit

# Assuming X and y are your prepared sequences from previous code
# ... (your existing code for data loading, preprocessing, and sequence creation)

# Define the number of splits for time series cross-validation
n_splits = 5  # You can adjust the number of splits

# Initialize TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=n_splits)

# Iterate through the splits
rmse_scores = []
for train_index, test_index in tscv.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Reshape the input data for RNN
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    # Build, compile, and train the model (same as before)
    model = Sequential()
    model.add(SimpleRNN(50, activation='relu', input_shape=(seq_length, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0) # Set verbose to 0 to suppress output

    # Make predictions and evaluate
    y_pred = model.predict(X_test)
    y_pred_original = scaler.inverse_transform(y_pred)
    y_test_original = scaler.inverse_transform(y_test.reshape(-1, 1))
    rmse = math.sqrt(mean_squared_error(y_test_original, y_pred_original))
    rmse_scores.append(rmse)

# Print the RMSE scores for each fold
print("RMSE scores for each fold:", rmse_scores)

# Calculate and print the mean and standard deviation of RMSE scores
mean_rmse = np.mean(rmse_scores)
std_rmse = np.std(rmse_scores)

print(f"\nMean RMSE: {mean_rmse}")
print(f"Standard Deviation of RMSE: {std_rmse}")
```
